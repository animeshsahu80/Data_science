```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import datasets

```

## Loading the dataset

```python
iris=datasets.load_iris()  #load the dataset


a=np.array(iris['data'])   
b=np.array(iris['target'])  #a and b have to be concatenated to be sent as an 'data' argument in p.Dataframe

new_data=np.column_stack((a,b))   #concatenated
col=np.char.add(iris['feature_names'],['target'])    #name of the coloumns
df=pd.DataFrame(new_data,columns=iris['feature_names']+['species'])  #final dataframe
```

```python
df['species'].value_counts()  #count of each species
   
print(df.species)
```

```python
df.species.replace([0,1,2], ['setosa', 'versicolor', 'virginica'], inplace=True)  # replacing target number with target names
df.head(10)
```

```python
sns.scatterplot(x='sepal length (cm)',y='sepal width (cm)',data=df,hue='species')
```

## Histogram, PDF, CDF

```python
df_setosa=df.loc[df['species']=='setosa']
df_virginica=df.loc[df['species']=='virginica']
df_versicolor=df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['petal length (cm)'],np.zeros_like(df_setosa['petal length (cm)']),'o')
plt.plot(df_virginica['petal length (cm)'],np.zeros_like(df_virginica['petal length (cm)']),'--')
plt.plot(df_versicolor['petal length (cm)'],np.zeros_like(df_versicolor['petal length (cm)']),'o')
plt.legend(['setosa','virginica','versicolor'])
plt.grid()
```

```python
g = sns.FacetGrid(df, hue="species",height=5)
g.map(sns.distplot,'petal length (cm)')
plt.legend(['setosa','versicolor','virginica'])
```

```python
g = sns.FacetGrid(df, hue="species",height=5)
g.map(sns.distplot,'petal width (cm)')
plt.legend(['setosa','virginica','versicolor'])
```

```python
g = sns.FacetGrid(df, hue="species",size=5)
g.map(sns.distplot,'sepal length (cm)')
plt.legend(['setosa','virginica','versicolor'])

```

```python
g = sns.FacetGrid(df, hue="species",size=5)
g.map(sns.distplot,'sepal width (cm)')
plt.legend(['setosa','virginica','versicolor'])
```

#### we can see here taht sepal width and sepal length are terrible at classification, and petal length is the best whereas petal width is in between 

```python
#setosa pdf and cdf

counts, bin_edges=np.histogram(df_setosa['petal length (cm)'],bins=10,density=True)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.grid()


#virginica pdf and cdf
counts, bin_edges=np.histogram(df_virginica['petal length (cm)'],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.grid()


#versicolor pdf and cdf
counts, bin_edges=np.histogram(df_versicolor['petal length (cm)'],bins=10)
pdf=counts/sum(counts)
cdf=np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
plt.grid()
```

## Mean, Variance and Std dev

```python
print('mean')
print('setosa--->',np.mean(df_setosa['petal length (cm)']))
print('virginica--->',np.mean(df_virginica['petal length (cm)']))
print('versicolor--->',np.mean(df_versicolor['petal length (cm)']))
```

```python
print('std dev')
print('setosa--->',np.std(df_setosa['petal length (cm)']))
print('virginica--->',np.std(df_virginica['petal length (cm)']))
print('versicolor--->',np.std(df_versicolor['petal length (cm)']))
```

 ## Medain, Percentile, Quantile, Medain absolute deviation

```python
print('median')
print('setosa--->',np.median(df_setosa['petal length (cm)']))
print('virginica--->',np.median(df_virginica['petal length (cm)']))
print('versicolor--->',np.median(df_versicolor['petal length (cm)']))
```

```python
print('percentile [0 25 50 75]')
print('setosa--->',np.percentile(df_setosa['petal length (cm)'],[0,25,50,75]))
print('virginica--->',np.percentile(df_virginica['petal length (cm)'],[0,25,50,75]))
print('versicolor--->',np.percentile(df_versicolor['petal length (cm)'],[0,25,50,75]))
```

```python
from statsmodels import robust

print('median absolute deviation')
print('setosa--->',robust.mad(df_setosa['petal length (cm)']))
print('virginica--->',robust.mad(df_virginica['petal length (cm)']))
print('versicolor--->',robust.mad(df_versicolor['petal length (cm)']))
```
